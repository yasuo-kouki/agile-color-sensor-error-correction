```python
import pandas as pd

data = [
    ["nomal_model_30", 16458, 69872, 11420, 716, 391.52, 608.52, 240.52, 319.37, 528.7, 417.726],
    ["nomal_model_40", 27885, 73008, 14544, 1177, 441, 462.7, 516.3, 421.78, 431.93, 454.742],
    ["nomal_model_50", 42273, 76928, 18464, 1655, 195.37, 366.37, 371.07, 224.67, 87.22, 248.94],
    ["nomal_model_60", 59560, 81648, 23184, 2399, 311.93, 582.3, 320.04, 593.63, 424.52, 446.484],
]

columns = ["filename", "bit", "sketch", "Global", "time", "1st", "2nd", "3rd", "4th", "5th", "mean"]

df = pd.DataFrame(data, columns=columns)

print(df)
```

```python
import matplotlib.pyplot as plt

# グラフの作成
plt.figure(figsize=(8, 5))
plt.bar(df["filename"], df["bit"], color="skyblue")
plt.xlabel("Model")
plt.ylabel("Bit")
plt.title("Bit Value by Model")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1st～5thの値をリスト化
data_to_plot = df[["1st", "2nd", "3rd", "4th", "5th"]].values

plt.figure(figsize=(10, 6))

colors = ["skyblue", "lightgreen", "lightcoral", "orange"]

# 箱ひげ図作成
box = plt.boxplot(data_to_plot.T, patch_artist=True, labels=df["filename"], widths=0.6,
                  boxprops=dict(facecolor='skyblue', color='black'),
                  medianprops=dict(color='red', linewidth=2),
                  whiskerprops=dict(color='black', linewidth=1.5),
                  capprops=dict(color='black', linewidth=1.5),
                  flierprops=dict(marker='o', color='purple', alpha=0.6))

# 箱ごとに色を変更
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# 平均値だけを重ねる
for i, y in enumerate(data_to_plot.T):
    plt.plot(i+1, np.mean(y), 'rD', markersize=8, label='Mean' if i==0 else "")

plt.title("1st~5th Values Distribution by Model", fontsize=14)
plt.ylabel("Value", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
