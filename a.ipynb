```python
import pandas as pd

# 抽出したいファイル名リスト
target_files = [
    "model_sparse0.005_csr_int8_param30",
    "model_sparse0.005_csr_int8_param70",
    "model_sparse0.005_csr_int8_param80",
    "model_sparse0.005_csr_int8_param90",
    "model_sparse0.005_csr_int8_param100",
    "model_sparse0.005_csr_int8_param200",
    "model_sparse0.005_csr_int8_param500",
    "model_sparse0.005_csr_int8_param700",
    "model_sparse0.005_csr_int8_param1000",
    "model_sparse0.005_csr_int8_param1100",
    "model_sparse0.005_csr_int8_param1200"
]

# 1st, 2nd, 3rd 列だけ抽出
df_selected = df[df["filename"].isin(target_files)][["filename", "1st", "2nd", "3rd"]].copy()

# インデックスをリセット
df_selected.reset_index(drop=True, inplace=True)

# 表示
print(df_selected)
```

```python
import matplotlib.pyplot as plt

# 1st～3rd の平均を計算して 'row_mean' 列を作成
df_selected["row_mean"] = df_selected[["1st", "2nd", "3rd"]].mean(axis=1)
# 確認
print(df_selected[["filename", "row_mean"]])

x = df_selected["filename"]
y = df_selected["row_mean"]

plt.figure(figsize=(14, 6))
plt.bar(x, y, color="skyblue")
plt.xticks(rotation=90, fontsize=8)
plt.xlabel("filename")
plt.ylabel("row mean of 1st~3rd")
plt.title("Row Mean of 1st/2nd/3rd for selected model_sparse0.005_csr_int8")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = df_selected["filename"]
y = df_selected["row_mean"]

plt.figure(figsize=(16, 6))
bars = plt.bar(x, y, color="skyblue", edgecolor="black")

# y軸グリッドを追加
plt.grid(axis="y", linestyle="--", alpha=0.7)

# 横軸ラベルを回転＆フォントサイズ調整
plt.xticks(rotation=45, fontsize=9, ha='right')
plt.yticks(fontsize=10)

plt.xlabel("Filename", fontsize=12)
plt.ylabel("Row Mean of 1st~3rd", fontsize=12)
plt.title("Average Performance (1st~3rd) for Selected model_sparse0.005_csr_int8", fontsize=14)

# 棒の上に値を表示
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 10, f'{height:.1f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
